```python
#W7A24
s = input("Nhập câu: ")
words = s.split()
if len(words) > 0:
    result = [words[0]] 
    for i in range(1, len(words)):
        if words[i] != words[i-1]:
            result.append(words[i])
    print(" ".join(result))
```

```python
#W7A29
arr = [int(x) for x in input("Nhập dãy số: ").split()]
if len(arr) == 0:
    print()
else:
    max_len = 1
    best_start = 0    
    current_len = 1
    current_start = 0    
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            current_len += 1
        else:
            if current_len > max_len:
                max_len = current_len
                best_start = current_start
            current_len = 1
            current_start = i  
    if current_len > max_len:
        max_len = current_len
        best_start = current_start
    print(arr[best_start : best_start + max_len])
```

```python
#W7A18
arr = [int(x) for x in input("Nhập dãy số: ").split()]
even_nums = [] 
odd_nums = []  
for x in arr:
    if x % 2 == 0:
        even_nums.append(x)
    else:
        odd_nums.append(x)
even_nums.sort()            
odd_nums.sort(reverse=True) 
result = even_nums + odd_nums
print(*result)
```

```python
#W7A26
s = input("Nhập chuỗi: ")
s += " " 
numbers = []
current_num_str = ""
for char in s:
    if char.isdigit():
        current_num_str += char
    else:
        if current_num_str != "":
            numbers.append(int(current_num_str))
            current_num_str = ""
if len(numbers) > 0:
    total = sum(numbers)
    avg = total / len(numbers)
    print(total, avg)
else:
    print(0, 0)
```

```python
#W7A23
s = input("Nhập câu: ")
for p in ",.?!;": 
    s = s.replace(p, " ")    
words = s.split()
count = 0
for w in words:
    if w.isalpha() and len(w) >= 2:
        if w[0].isupper() and w[1:].islower():
            count += 1
print(count)
```

```python
#W7A21
d = int(input("Nhập ngày: "))
m = int(input("Nhập tháng: "))
days_in_months = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
if m < 1 or m > 12 or d < 1 or d > days_in_months[m-1]:
    print("Invalid")
else:
    day_count = 0
    for i in range(m - 1):
        day_count += days_in_months[i]
    day_count += d
    print(day_count)
```

```python
#W7A27
s = input("Nhập dãy ngoặc: ")
stack = []
mapping = {')': '(', ']': '[', '}': '{'}
is_valid = True
for char in s:
    if char in "({[":
        stack.append(char) 
    elif char in ")}]":
        if len(stack) == 0:
            is_valid = False
            break
        last_open = stack.pop() 
        if mapping[char] != last_open:
            is_valid = False
            break
if is_valid and len(stack) == 0:
    print("YES")
else:
    print("NO")
```

```python
#W7A22
try:
    n = int(input("Nhập số lượng sinh viên: "))
    name_counts = {}  
    for _ in range(n):
        full_name = input()
        first_name = full_name.split()[-1]
        
        if first_name in name_counts:
            name_counts[first_name] += 1
        else:
            name_counts[first_name] = 1
    max_count = -1
    best_name = ""
    sorted_names = sorted(name_counts.keys())    
    for name in sorted_names:
        if name_counts[name] > max_count:
            max_count = name_counts[name]
            best_name = name            
    print(best_name)
except:
    pass
```

```python
#W7A16
arr = [int(x) for x in input("Nhập dãy số: ").split()]
counts = {}
for x in arr:
    if x in counts:
        counts[x] += 1
    else:
        counts[x] = 1
delete_count = 0
for x in counts:
    c = counts[x]
    if c < x:
        delete_count += c
    elif c > x:
        delete_count += (c - x)
if delete_count == 0:
    print("YES")
else:
    print("NO", delete_count)
```
