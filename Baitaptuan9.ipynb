```python
#W8A1
import math
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def surface_area(self):
        return 2 * 3.14 * self.r * (self.r + self.h)    
    def volume(self):
        return 3.14 * (self.r ** 2) * self.h
try:
    line = input("Nhập bán kính và chiều cao (r h): ")
    r, h = map(float, line.split())    
    cyl = Cylinder(r, h)
    print(f"{cyl.surface_area():.2f} {cyl.volume():.2f}")
except:
    pass
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return (self.w + self.h) * 2
    def scale(self, k):
        self.w = self.w * k
        self.h = self.h * k
try:
    w, h, k = map(float, input("Nhập w, h và hệ số k: ").split())    
    rect = Rectangle(w, h)
    rect.scale(k)
    print(f"{rect.area():.2f} {rect.perimeter():.2f}")
except:
    pass
```

```python
#W8A3
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y        
    def check_position(self):
        if self.x == 0 and self.y == 0:
            print("Gốc tọa độ")
        elif self.x == 0:
            print("Trục tung")
        elif self.y == 0:
            print("Trục hoành")
        else:
            print("Không nằm trên trục nào")            
    def distance_to_origin(self):
        return math.sqrt(self.x**2 + self.y**2)
try:
    x, y = map(float, input("Nhập tọa độ x y: ").split())
    p = Point(x, y)    
    p.check_position()
    print(f"{p.distance_to_origin():.2f}")
except:
    pass
```

```python
#W8A13
import math
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def get_area(self):
        if (self.a + self.b > self.c) and (self.a + self.c > self.b) and (self.b + self.c > self.a):
            p = (self.a + self.b + self.c) / 2
            area = math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
            return area
        else:
            return -1
try:
    a, b, c = map(int, input("Nhập 3 cạnh: ").split())
    t = Triangle(a, b, c)
    result = t.get_area()
    if result == -1:
        print("invalid")
    else:
        print(f"{result:.2f}")
except:
    pass
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance) 
    def deposit(self, amount):
        self.balance += amount
        
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
try:
    line1 = input().split()
    owner = line1[0]
    initial_balance = float(line1[1])    
    tk = BankAccount(owner, initial_balance)    
    n = int(input())
    for _ in range(n):
        cmd = input().split()
        action = cmd[0]
        money = float(cmd[1])        
        if action == "DEPOSIT":
            tk.deposit(money)
        elif action == "WITHDRAW":
            tk.withdraw(money)            
    print(f"{tk.balance:.2f}")
except:
    pass
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []      
    def add_score(self, score):
        self.scores.append(score)        
    def avg(self):
        if len(self.scores) == 0: return 0
        return sum(self.scores) / len(self.scores)    
    def rank(self):
        a = self.avg()
        if a >= 8.0: return "Excellent"
        if a >= 6.5: return "Good"
        if a >= 5.0: return "Average"
        return "Poor"
try:
    name = input()
    st = Student(name)
    n = int(input())    
    for _ in range(n):
        parts = input().split()
        score = float(parts[-1]) 
        st.add_score(score)       
    print(f"{st.name} {st.avg():.2f} {st.rank()}")
except:
    pass
```

```python
#W8A4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def add(self): return self.a + self.b
    def subtract(self): return self.a - self.b
    def product(self): return self.a * self.b
    def divide(self): 
        if self.b == 0: return "Error"
        return self.a / self.b
    def power(self): return self.a ** self.b
try:
    a, b = map(float, input("Nhập 2 số: ").split())
    calc = Calculator(a, b)
    print("Cộng:", calc.add())
    print("Trừ:", calc.subtract())
    print("Nhân:", calc.product())
    print("Chia:", calc.divide())
except:
    pass
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = [] 
    def add_item(self, name, price):
        item = {"name": name, "price": price}
        self.items.append(item)        
    def get_total(self):
        total = 0
        for item in self.items:
            total += item["price"]
        return total
        
    def show_cart(self):
        for item in self.items:
            print(f"{item['name']}: {item['price']}")
cart = ShoppingCart()
print("Nhập tên và giá (vd: Tao 10), gõ 'exit' để dừng:")
while True:
    line = input()
    if line == "exit": break
    try:
        parts = line.split()
        price = float(parts[-1])
        name = " ".join(parts[:-1])
        cart.add_item(name, price)
    except:
        continue
cart.show_cart()
print("Tổng tiền:", cart.get_total())
```
