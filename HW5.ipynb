```python
#W5A2
def swap(x,y):
    return [y,x]

a = int(input(""))
b = int(input(""))
print(a,b)
a,b = swap(a,b)
print(a,b)
```

```python
#W5A4
def perfect(x):
    sum = 0
    for i in range(1,x//2+1):
        if x%i == 0:
            sum+=i
    return sum == i

x = int(input(""))
if perfect(x):
    print("Yes")
else:
    print("No") 
```

```python
#W5A5
def first_index(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

n = list(int(i) for i in input().split())
p = int(input(""))
print(first_index(n, p))
```

```python
#W5A6
def gt(a):
    if a == 1:
        return 1
    return a*gt(a-1)

print(gt(5))
```
