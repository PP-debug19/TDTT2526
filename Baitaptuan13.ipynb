```python
#W11A1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        n = f.read().strip()
    return int(n[::-1])
print(reverse('input1.txt'))
```

```python
#W11A2
def canWinNim(path: str) -> bool:
    with open(path, 'r') as f:
        n = int(f.read().strip())
    return n % 4 != 0
print(canWinNim('input2.txt'))
```

```python
#W11A3
def housesOfHogwarts(path: str):
    with open(path, 'r') as f:
        lines = f.readlines()
    n = int(lines[0].strip())
    for i in range(1, n+1):
        print(lines[i].strip())
        
housesOfHogwarts('input3.txt')
```

```python
#W11A4
def areAnagrams(path1: str, path2: str) -> bool:
    def clean(s):
        import re
        return ''.join(sorted(re.sub(r'[^a-z]', '', s.lower())))
    with open(path1) as f1, open(path2) as f2:
        s1, s2 = f1.read(), f2.read()
    return clean(s1) == clean(s2)

print(areAnagrams('input4.txt','input4.1.txt'))
```

```python
#W11A9
import csv
def grade10(path: str):
    with open(path) as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row['THCS4'] == '10':
                print(row['Ho'] + " " + row['Ten'])

grade10('input9.txt')
```

```python
#W11A5
def count(path: str, character: str) -> int:
    with open(path) as f:
        text = f.read()
    return text.count(character)

print(count('input5.txt', 'e'))
```

```python
#W11A6
def countCharacter(path: str, letter: str) -> int:
    with open(path) as f:
        text = f.read().lower()
    return text.count(letter.lower())

print(countCharacter('input6.txt', 'e'))
```

```python
#W11A7
def checkFile(path: str):
    try:
        f = open(path, 'r')
        print("YES")
        f.close()
    except:
        print("NO")

checkFile('input7.txt')
```

```python
#W11A11
def minMax(path: str):
    try:
        with open(path) as f:
            nums = list(map(int, f.read().split()))
        print(min(nums), max(nums))
    except:
        print("Mission failed")

minMax('input11.txt')
```

```python
#W11A13
def moveZeroes(fileName: str) -> list[int]:
    with open(fileName) as f:
        f.readline()
        nums = list(map(int, f.read().split()))
    res = [x for x in nums if x != 0] + [0]*nums.count(0)
    return res

print(moveZeroes('input13.txt'))
```
