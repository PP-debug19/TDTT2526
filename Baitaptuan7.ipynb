```python
#W7A2
def count_occurrences(arr, x):
    count = 0
    for num in arr:
        if num == x:
            count += 1
    return count

input_list = [1, 3, 4, 60, 70, 90, 100, 60, 60, 60]
x_value = 60
print(count_occurrences(input_list, x_value))
```

```python
def find_nemo_neighbors():
    n = int(input())
    
    names = []
    for i in range(n):
        name = input()
        names.append(name)
    
    nemo_index = -1
    for i in range(n):
        if names[i] == "Nemo":
            nemo_index = i
            break
    
    if nemo_index != -1:
        nguoi_truoc = names[nemo_index - 1]
        nguoi_sau = names[nemo_index + 1]
        print(f"{nguoi_truoc} and {nguoi_sau}")

if __name__ == "__main__":
    find_nemo_neighbors()
```

```python
#W7A14
def find_sevens():
    try:
        line1 = input()
        if not line1: return 
        n = int(line1)
        line2 = input()
        str_nums = line2.split()
        arr = []
        for s in str_nums:
            arr.append(int(s))
        found_indices = []
        for i in range(len(arr)):
            if arr[i] == 7:
                found_indices.append(i)
        if len(found_indices) == 0:
            print("Not found")
        else:
            for i in range(len(found_indices) - 1, -1, -1):
                print(found_indices[i], end=" ")
            print()            
    except ValueError:
        pass

if __name__ == "__main__":
    find_sevens()
```

```python
#W7A1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2  
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1      
    return -1 

arr_test = [1, 3, 4, 60, 70, 90, 100, 1000]
print(binary_search(arr_test, 0, len(arr_test) - 1, 60)) 
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp
    return arr

ds = [60, 3, 1, 70, 4, 100, 90]
print(bubble_sort(ds))
```

```python
#W7A4
def most_frequent(arr):
    counts = {} 
    for num in arr:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1
    max_count = 0
    for key in counts:
        if counts[key] > max_count:
            max_count = counts[key]
    for num in arr:
        if counts[num] == max_count:
            return num, max_count

input_arr = [2, 3, 4, 2, 2, 5, 1, 1, 1] 
print(most_frequent(input_arr))
```

```python
#W7A5
def count_pairs_with_sum(arr, X):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                count += 1
    return count

lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]
x_val = 10
print(count_pairs_with_sum(lst, x_val))
```

```python
#W7A6
def flatten_list(nested_list):
    flat_result = []
    for sublist in nested_list:
        for item in sublist:
            flat_result.append(item)       
    return flat_result

input_nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print(flatten_list(input_nested))
```

```python
#W7A9
def count_smaller_on_right(nums):
    result = []
    n = len(nums)
    for i in range(n):
        count = 0
        for j in range(i + 1, n):
            if nums[j] < nums[i]:
                count += 1
        result.append(count) 
    return result

inp = [5, 2, 6, 1]
print(count_smaller_on_right(inp))
```
