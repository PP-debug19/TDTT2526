```python
#Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
#...} (đếm số >0, <0, =0)
lst=list(map(int,input().split()))
pos=0
neg=0
zeros=0
for x in lst:
    if x>0:
        pos+=1
    elif x<0:
        neg+=1
    else:
        zeros+=1
print({'positives':pos,'negatives':neg,'zeros':zeros})
```

```python
#Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng
#chuỗi và lưu vào dictionary. 
strings=input().split()
count={}
for x in strings:
    if x not in count:
        count[x]=0
    count[x]+=1
print(count)
```

```python
#Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
#trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
#trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
#key. 
# Nhập dict1 và dict2
dict1 = eval(input("Nhập dict1: "))
dict2 = eval(input("Nhập dict2: "))

# Tạo dict kết quả rỗng
merged = {}

# Duyệt từng phần tử trong dict1
for k in dict1:
    if k in merged:          # nếu key đã có thì cộng thêm
        merged[k] = merged[k] + dict1[k]
    else:                    # nếu chưa có thì gán mới
        merged[k] = dict1[k]

# Duyệt từng phần tử trong dict2
for k in dict2:
    if k in merged:
        merged[k] = merged[k] + dict2[k]
    else:
        merged[k] = dict2[k]

# Sắp xếp theo thứ tự tăng dần của key (thủ công hơn)
keys = list(merged.keys())
keys.sort()   # sắp xếp list key

# Tạo dict mới theo thứ tự đã sắp
sorted_dict = {}
for k in keys:
    sorted_dict[k] = merged[k]

print(sorted_dict)
```

```python
#[Tách số chẵn và lẻ từ tuple]
#Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
#tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
#thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt. 
lst=tuple(map(int,input().split()))
chan=[]
le=[]
for x in lst:
    if x%2==0:
        chan.append(x)
    else:
        le.append(x)
chan_tup=tuple(chan)
le_tup=tuple(le)
print(chan_tup)
print(le_tup)
```

```python
#[Tìm phần tử xuất hiện nhiều nhất]
#Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
#Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
#trong số đó. 
# Nhập list số nguyên
nums = list(map(int, input("Nhập list số nguyên: ").split()))

counts = {}
for n in nums:
    if n in counts:
        counts[n] += 1
    else:
        counts[n] = 1

max_count = max(counts.values())

candidates = []
for num, cnt in counts.items():
    if cnt == max_count:
        candidates.append(num)

print(min(candidates))
```

```python
#[Lọc dictionary theo giá trị]
#Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
#nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
#keyvalue từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới. 
d=input()
k=int(input())
loc={}
for key in d:
    if d[key]>k:
        loc[key]=d[key]
print(loc)
```
